```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import beta, binom

sns.set(style='darkgrid', font_scale=1.5)
```

```python
# prior parameters on theta
A = 2
B = 2

# number of simulations, number of trials w/in sims
n_sim = 1000
n_bin = 250

# posterior mode
def beta_mode(a=2, b=2):
    return (a-1)/(a+b-2)

theta_true = np.zeros(n_sim)
theta_mode = np.zeros(n_sim)
theta_mean = np.zeros(n_sim)
theta_median = np.zeros(n_sim)

stop_early = np.zeros(n_sim, dtype=bool)
stop_trial = np.zeros(n_sim, dtype=int)

crit = .9
theta_comp = 0.25
```

```python
for s_i in range(n_sim):
    A = B = 2
    theta_t = beta.rvs(a=A, b=B)
    theta_true[s_i] = theta_t
    data_t = binom(n=1, p=theta_t).rvs(size=n_bin)
    for d_i, d_t in enumerate(data_t):
        A += d_t
        B += 1-d_t
        beta_update = beta(a=A, b=B)
        quant_t = beta_update.ppf(1-crit)
        if quant_t > theta_comp:
            stop_early[s_i] = True
            break
    stop_trial[s_i] = d_i
    theta_mode[s_i] = beta_mode(a=A, b=B)
    theta_mean[s_i] = beta_update.mean()
    theta_median[s_i] = beta_update.median()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,8))
ax_stop, ax_fin = axes

ax_stop.plot([0,1],[0,1],'k:')
ax_stop.plot(theta_true[stop_early], theta_mode[stop_early], 'o')
ax_stop.plot(theta_true[stop_early], theta_mean[stop_early], 's')
ax_stop.plot(theta_true[stop_early], theta_median[stop_early], '^')
ax_stop.set(xlim=(-0.025,1.025), ylim=(-0.025,1.025), xlabel=r'$\theta$', ylabel=r'$\hat{\theta}$')

ax_fin.plot(theta_true[np.invert(stop_early)], theta_mode[np.invert(stop_early)], 'o')
ax_fin.plot(theta_true[np.invert(stop_early)], theta_mean[np.invert(stop_early)], 's')
ax_fin.plot(theta_true[np.invert(stop_early)], theta_median[np.invert(stop_early)], '^')
ax_fin.set(xlim=(-0.025,1.025), ylim=(-0.025,1.025), xlabel=r'$\theta$', ylabel=r'$\hat{\theta}$');
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
hh = ax.hist(stop_trial, bins=np.arange(0,251,10), histtype='step', lw=3)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
hh = ax.hist(theta_mode[stop_early]-theta_true[stop_early], bins=np.linspace(-1,1,21), histtype='step', lw=3)
hh = ax.hist(theta_mean[stop_early]-theta_true[stop_early], bins=np.linspace(-1,1,21), histtype='step', lw=3)
hh = ax.hist(theta_median[stop_early]-theta_true[stop_early], bins=np.linspace(-1,1,21), histtype='step', lw=3)
print(np.mean(theta_mode[stop_early]-theta_true[stop_early]),
      np.mean(theta_mean[stop_early]-theta_true[stop_early]),
      np.mean(theta_median[stop_early]-theta_true[stop_early]))
```
